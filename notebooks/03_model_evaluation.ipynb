```python
import joblib
import os
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
model_path = os.path.join("..", "outputs", "models", "random_forest_model.pkl")

# Load model
rf_model_path = r"C:\Users\Ritik\Desktop\credit score\outputs\models\random_forest_model.pkl"
rf_model = joblib.load(rf_model_path)
xgb_model_path = r"C:\Users\Ritik\Desktop\credit score\outputs\models\xgboost_model.pkl"
xgb_model = joblib.load(xgb_model_path)
```

```python
features_to_use = [
        'repayment_efficiency', 'total_repay', 'num_repay', 'deposit_frequency',
        'borrow_to_deposit_ratio', 'total_deposit', 'net_borrow', 'activity_score'
    ]
```

```python
importances = xgb_model.feature_importances_

# Create DataFrame for plotting
feat_imp_df = pd.DataFrame({
    'Feature': features_to_use,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)
plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=feat_imp_df, palette='viridis')
plt.title("Feature Importances from XGBoost Model", fontsize=16)
plt.xlabel("Importance", fontsize=14)
plt.ylabel("Feature", fontsize=14)
plt.tight_layout()
plt.show()
```
